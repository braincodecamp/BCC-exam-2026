```python
# โค้ดในช่องนี้เป็นโค้ดสำหรับการดาวน์โหลดและแสดง (plot) ข้อมูล ท่านไม่จำเป็นต้องแก้ไขโค้ดในช่องนี้

import io
import requests
import numpy as np
import plotly.graph_objects as go

# ── 1. Download Data ──────────────────────────────────────────────────────────

DATA_URL = (
    f"https://raw.githubusercontent.com/braincodecamp"
    "/BCC-exam-2026/main/part-5/organ.npy"
)

response = requests.get(DATA_URL)
response.raise_for_status()

points = np.load(io.BytesIO(response.content))

# ── 2. Dimensions ─────────────────────────────────────────────────────────────

print(f"Shape  : {points.shape}")
print(f"Dtype  : {points.dtype}")

# ── 3. Helper: interactive 3D scatter (Plotly) ────────────────────────────────

def plot_point_cloud(pts, title=None):
    fig = go.Figure(
        go.Scatter3d(
            x=pts[:, 0], y=pts[:, 1], z=pts[:, 2],
            mode="markers",
        )
    )
    fig.update_layout(
        title=dict(text=title, font=dict(size=14)),
        scene=dict(
            xaxis_title="X",
            yaxis_title="Y",
            zaxis_title="Z",
        ),
    )
    fig.show()
```

```python
# ข้อมูลพิกัดของอวัยวะจะอยู่ในตัวแปร points

# ฟังก์ชั่น plot_point_cloud() เป็นฟังก์ชั่นที่ใช้ในการแสดงข้อมูลใน 3 มิติ โดยท่านสามารถหมุนกราฟไปมา
# เพื่อดูข้อมูลรอบด้านได้

plot_point_cloud(points, "ข้อมูลเริ่มต้น")
```
